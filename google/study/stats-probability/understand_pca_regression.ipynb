```python
import numpy as np
# 3x3 design matrix
X = np.array([
    [1, 2, 3],
    [1, 4, 5],
    [1, 6, 7],
])
# Projection (hat) matrix: H = X (X'X)^{-1} X'
H = X @ np.linalg.inv(X.T @ X) @ X.T
```

```python
H.shape
```

```python

y = np.array([1, 2, 3])
y_hat = H @ y  # = X @ beta_hat
```
